```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Variables para DataFrame
titles = []
companies_name = []
companies_link = []
companies_calcifications = []
companies_locations = []
how_longs = []
short_descriptions = []

# Captura del tema de interés de la persona
interest = input("¿Cuál es el trabajo de tu interés?: ")
print('\n')

# Acceso a la página
driver =  webdriver.Chrome(service = Service(ChromeDriverManager().install()))
driver.get("https://co.computrabajo.com/")
search_bar = driver.find_element(By.CSS_SELECTOR, "input#prof-cat-search-input")
search_bar.clear()
search_bar.send_keys(interest)
search_bar.send_keys(Keys.RETURN)

# Título del trabajo
title_jobs = driver.find_elements(By.XPATH, "//a[@class='js-o-link fc_base']")
# title_jobs = [title.text for title in title_jobs]

# Sub título (Empresa - Ubicación - Calificación)
sub_titles = driver.find_elements(By.CSS_SELECTOR, "p.fs16.fc_base.mt5.mb10")

# Ciclo para separa cada uno de los elementos
for sub_title in sub_titles:
    
    # Nombre de la empresa
    if sub_title.find_elements(By.TAG_NAME, 'a'): 
        companies_name.append(sub_title.find_element(By.TAG_NAME, 'a').text)
    else: companies_name.append("Sin nombre de empresa...")
        
    # Enlase de la empresa
    if sub_title.find_elements(By.TAG_NAME, 'a'): 
        companies_link.append(sub_title.find_element(By.TAG_NAME, 'a').get_attribute('href'))
    else: companies_link.append("Sin enlace de empresa...")
        
    # Calificación de la empresa
    if sub_title.find_elements(By.CSS_SELECTOR, 'span.ml10.mr10'): 
        companies_calcifications.append(sub_title.find_element(By.CSS_SELECTOR, 'span.ml10.mr10').text)
    else: companies_calcifications.append("Sin calificación de empresa...")
        
print(companies_name,'\n\n\n' ,companies_link,'\n\n\n' ,companies_calification)

# Variable para poder hacer el ciclo en las diferentes páginas
# next_button_validate = True

# while(next_button_validate):
#     next_button = driver.find_element(By.CSS_SELECTOR, 'span.b_next.buildLink.cp')


# pagination = driver.find_element(By.XPATH ,"//li[@class='andes-pagination__page-count']").text
# pagination = [ int(s) for s in pagination.split() if s.isdigit()][0]

# records = []

# for _product in range(1,pagination+1):
#     if _product != pagination:
#         next_page_button = driver.find_element_by_css_selector("a[title='Siguiente']")

#     #title_products = driver.find_elements_by_xpath("//h2[@class='ui-search-item__title']")
#     title_jobs = driver.find_elements(By.XPATH,"//a[@class='js-o-link fc_base']']")
#     title_jobs = [title.text for title in title_products]


#     #price_products = driver.find_elements_by_xpath("//li[@class='ui-search-layout__item']//div[@class='ui-search-result__content-columns']//div[@class='ui-search-result__content-column ui-search-result__content-column--left']/div[1]/div//div[@class='ui-search-price__second-line']//span[@class='price-tag-amount']//span[2]")
#     price_products = driver.find_elements_by_xpath("//div[@class='ui-search-price__second-line']//span[@class='price-tag ui-search-price__part']//span[@class='price-tag-amount']//span[2]")
#     price_products = [ price.text for price in price_products   ]


#     #links_products = driver.find_elements_by_xpath("//div[@class='ui-search-item__group ui-search-item__group--title']//a[1]")
#     links_products = driver.find_elements_by_xpath("//div[@class='ui-search-result__wrapper']//div[@class='ui-search-result__image']//a[@class='ui-search-link']")
#     links_products = [ link.get_attribute("href") for link in links_products   ]
   
#     data_products = {

#         "name_product":title_products,
#         "price_product":price_products,
#         "link_product":links_products

#     }
    

#     df =  pd.DataFrame(data_products)
#     records.append(df)
#     if _product != pagination:
#         driver.execute_script("arguments[0].click()", next_page_button)
#     time.sleep(4)
    
# df = pd.concat(records)
# df.to_csv("PRODUCTOS.csv")
# time.sleep(4)
driver.close()
```
