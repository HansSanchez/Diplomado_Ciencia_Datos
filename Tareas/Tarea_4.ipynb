# Título: 
<p>Diseño de una plataforma prototipo capaz de hacer web scraping en CompuTrabajo para análisis de ofertas laborales en Colombia.</p>

# Reporte No. 1
<p>
Los registros actuales revelan un porcentaje significativo de personas que están desempleadas en Colombia, lo anterior debido a diferentes factores de vida, entre ellos lo son: el no tener la posibilidad de acceder a una herramienta de tecnología para consulta, desconocimiento de dispositivos o desconocimiento de la plataforma en estudio.
    <br>
    <br>
Por lo tanto, se busca tomar los datos de la plataforma CompuTrabajo que permitan establecer una puntuación relevante de las mejores ofertas laborales dependiendo algunas variables que puedan ser interesantes para las personas. 
    <br>
    <br>
En conclusión, la investigación contribuirá a ampliar la información a nivel general para contrastar ofertas laborales y poder generar una puntuación de relevancia según el interés de la personas que estén consultando el análisis recolectado.

</p>

# Descripción del problema
<p>
    La tasa de desempleo en Colombia, es decir, de aquellas personas que están en una condición de “desocupados”
    es del 10,7% de la población, a septiembre del 2022 <b>(DANE, 2022).</b>
    <br>
    <br>
    Es decir, un total de 2.600.000 personas en el territorio colombiano actualmente no cuentan con un trabajo
    para suplir sus necesidades o las de su familia (DANE, 2022).
    <br>
    <br>
    Además, las personas que buscan trabajo presentan una mayor dificultad para conocer ofertas laborales, tal vez
    por falta de herramientas y/o desconocimiento general (DANE, 2022).
    <br>
    <br>
    Por eso, este proyecto tiene como objetivo implementar las tecnologías y herramientas de la ciencia de datos
    para diseñar y desarrollar una plataforma prototipo que puntúe y notifique a nivel masivo nuevas ofertas de
    trabajo acerca de un tema de interés.
    <br>
    <br>
    <ol>
        <li>
            ¿Cómo puntuar las mejores ofertas laborales mediante la ciencia de datos?
        </li>
    </ol>
</p>
